```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_sm, y_sm, test_size=0.70,stratify=y_sm, random_state=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print(df['Churn'].value_counts())
sns.countplot(x='Churn', data=df, palette='coolwarm')
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(max_iter=1000)
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
from sklearn.metrics import accuracy_score
y_real=reg.predict(X_test)
accuracy=accuracy_score(y_real,y_test)
from sklearn.metrics import classification_report
clf_rep=classification_report(y_test,y_real)
pd.DataFrame(classification_report(y_test, y_real, output_dict=True)).T


```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_real)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Churn", "Churn"])
disp.plot(cmap="coolwarm")
plt.title('Confusion Matrix')
plt.show()
```
